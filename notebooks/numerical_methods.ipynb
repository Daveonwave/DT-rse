## Finite Differences Method
#### Backward difference

```python
def f(x):
    y = np.sin(x) + x + x * np.sin(x)
    return y

def f_prime(x):
    y = np.sin(x) + (x + 1) * np.cos(x) + 1
    return y
```

```python
x = np.linspace(-30, 30, 1500)

plt.rcParams["figure.figsize"] = [7.50, 3.50]
plt.rcParams["figure.autolayout"] = True

plt.plot(x, f(x), color='red')
plt.plot(x, f_prime(x), color='blue')

plt.show()
```

```python
def fdm(x, f, dt):
    i = 0
    y = []
    x_ = []
    
    while i < 200:
        dx = (f(x + dt) - f(x)) / dt
        y.append(dx)
        x_.append(x)
        x = x + dt
        i += 1
    return x_, y
```

```python
x = np.linspace(-10, 10, 1500)
dt = 1

plt.rcParams["figure.figsize"] = [7.50, 3.50]
plt.rcParams["figure.autolayout"] = True

plt.plot(x, f_prime(x), color='blue')

x_, y_ = fdm(x[0], f, dt)
plt.plot(x_[:20], y_[:20], color='green')

plt.show()
```

```python
err = [abs(y_[i] - f(x[i])) for i in range(len(x_))]

plt.plot(err)
```
